# Step 2 — Create First Flask App in Jupyter

| Method | Endpoint    | Purpose           |
| ------ | ----------- | ----------------- |
| GET    | /           | Check API running |
| GET    | /users      | Get all users     |
| GET    | /users/<id> | Get single user   |
| POST   | /users      | Create user       |
| PUT    | /users/<id> | Update user       |
| DELETE | /users/<id> | Delete user       |


```python
# Import Flask class
from flask import Flask

# Create Flask application object
# __name__ tells Flask where the app is located
app = Flask(__name__)

# Create first route (URL endpoint)
@app.route('/')
def home():
    """
    This function runs when user visits:
    http://127.0.0.1:5000/
    """
    
    return "Flask API is working successfully!"


# Function to run Flask app
def run_app():
    """
    This function starts Flask server
    """
    
    app.run(port=5000)
```

### MODULE 3 — Run Flask in Jupyter

```python
# Import threading module
# Thread allows Flask to run in background

from threading import Thread

# Create new thread
thread = Thread(target=run_app)

# Start Flask server
thread.start()
```

Step 8 — Create GET /users API

```python
from flask import Flask, jsonify

app = Flask(__name__)

# Sample in-memory database
# This acts like a temporary database

# Sample data
users = [
    {"id": 1, "name": "Aparna", "age": 25},
    {"id": 2, "name": "Rahul", "age": 30}
]


@app.route('/')
def home():
    """
    Home route
    Used to check API is working
    """
    
    return "Flask API is working!"


@app.route('/users', methods=['GET'])
def get_users():
    """
    GET API
    Returns list of all users
    """
    
    # jsonify converts Python list to JSON
    return jsonify(users)


@app.route('/users/<int:user_id>', methods=['GET'])
def get_user(user_id):
    """
    GET API to fetch single user
    using user ID
    """
    
    # Loop through users
    for user in users:
        
        # Check if ID matches
        if user["id"] == user_id:
            
            return jsonify(user)
    
    
    # If user not found
    return jsonify({"message": "User not found"})

from flask import request

@app.route('/users', methods=['POST'])
def add_user():
    """
    POST API
    Used to add new user
    """
    
    # Get JSON data from request
    new_user = request.get_json()
    
    
    # Add new user to list
    users.append(new_user)
    
    
    return jsonify({
        "message": "User added successfully",
        "user": new_user
    })

@app.route('/users/<int:user_id>', methods=['PUT'])
def update_user(user_id):
    """
    PUT API
    Updates existing user
    """
    
    updated_data = request.get_json()
    
    
    for user in users:
        
        if user["id"] == user_id:
            
            user.update(updated_data)
            
            return jsonify({
                "message": "User updated",
                "user": user
            })
    
    
    return jsonify({
        "message": "User not found"
    })

@app.route('/users/<int:user_id>', methods=['DELETE'])
def delete_user(user_id):
    """
    DELETE API
    Removes user
    """
    
    for user in users:
        
        if user["id"] == user_id:
            
            users.remove(user)
            
            return jsonify({
                "message": "User deleted"
            })
    
    
    return jsonify({
        "message": "User not found"
    })
    
def run_app():
    app.run(port=5000)
```

Step 9 — Restart Flask Server

Stop old thread (Kernel → Restart)

```python
from threading import Thread

thread = Thread(target=run_app)
thread.start()
```
